```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# For encoding categorical values
from sklearn.preprocessing import LabelEncoder

# For splitting data into train and test sets
from sklearn.model_selection import train_test_split

# For creating a decision tree model
from sklearn.tree import DecisionTreeClassifier

# For evaluating model performance
from sklearn.metrics import classification_report, confusion_matrix

# For visualizing the decision tree
from sklearn.tree import plot_tree
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum() #checking if there is null values
```

```python
df.sex.unique()
```

```python
pd.get_dummies(df['sex']).head()
```

```python
sex = pd.get_dummies(df['sex'], drop_first=True)
sex.head()
```

```python
pd.get_dummies(df['island']).head()
```

```python
island = pd.get_dummies(df['island'], drop_first=True)
island.head()
```

```python
new_data = pd.concat([df,island,sex], axis=1)
```

```python
new_data.head()
```

```python
new_data.drop(['island', 'sex'], axis=1, inplace=True)
```

```python
Y = new_data.species
Y.head()
```

```python
Y = Y.map({'Adelie':0, 'Chinstrap':1, 'Gentoo':2})
Y.head()
```

```python
new_data.drop('species', axis=1, inplace=True)
```

```python
X=new_data
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators= 5, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
